# Visualization of Results
Convergence plots, metric comparisons, and performance analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import os

# Set up paths relative to this notebook
NOTEBOOK_DIR = os.getcwd()
PROJECT_DIR = os.path.dirname(NOTEBOOK_DIR)
MODEL_RESULTS_PATH = os.path.join(PROJECT_DIR, 'results', 'model_results.csv')
WOA_DATA_PATH = os.path.join(PROJECT_DIR, 'models', 'woa_results.json')
FIGURES_DIR = os.path.join(PROJECT_DIR, 'figures')

if not os.path.exists(FIGURES_DIR):
    os.makedirs(FIGURES_DIR)
```

```python
if os.path.exists(MODEL_RESULTS_PATH):
    results = pd.read_csv(MODEL_RESULTS_PATH)
else:
    print("Results CSV not found. Please run 02_train_all_models.ipynb first.")
    results = None
```

```python
if results is not None:
    plt.figure(figsize=(10, 6))
    plt.bar(results["Model"], results["RMSE"], color=['gray', 'blue', 'orange'])
    plt.title("RMSE Comparison Between Models")
    plt.ylabel("RMSE")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig(os.path.join(FIGURES_DIR, "rmse_comparison.png"))
    plt.show()
```

```python
if results is not None:
    plt.figure(figsize=(10, 6))
    plt.bar(results["Model"], results["Time"], color=['gray', 'blue', 'orange'])
    plt.title("Training Time Comparison")
    plt.ylabel("Time (seconds)")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig(os.path.join(FIGURES_DIR, "training_time.png"))
    plt.show()
```

```python
if os.path.exists(WOA_DATA_PATH):
    with open(WOA_DATA_PATH, "r") as f:
        woa_data = json.load(f)
    
    history = woa_data.get("history", [])
    if history:
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(history) + 1), history, marker='o', color='orange')
        plt.title("WOA Convergence Plot")
        plt.xlabel("Iteration")
        plt.ylabel("Best Fitness (RMSE)")
        plt.grid(True)
        plt.savefig(os.path.join(FIGURES_DIR, "woa_convergence.png"))
        plt.show()
```
